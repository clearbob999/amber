```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

myfont = FontProperties(family='Times New Roman', weight='light', size=9)

fig, ax = plt.subplots(figsize=(8, 6))

# 数据
start_point = 0.2
spacing = 0.8
num_points = 15
x1 = [start_point + i * spacing for i in range(num_points)]
y1 = [2.25, 2.09, 1.84, 1.73, 0.95, 0.94, 0.8, 0.66, 0.45, 0.22, 0.19, -0.43, -0.03, -0.21, -2.03]

x2 = [start_point + 0.2 + i * spacing for i in range(num_points)]
y2 = [2.3, 2.44, 1.94, 1.92, 0.92, 0.77, 1.17, 0.19, 0.62, -0.13, 0.42, -0.65, 0.44, -0.34, -1.11]

# 绘制条形图
ax.bar(x1, y1, width=0.2, color='r', label='AMBER2')
ax.bar(x2, y2, width=0.2, color='k', label='AMBER')

# 设置x轴刻度标签和字体大小
x_labels = ['H223A', 'W210A', 'I4A', 'L207A', 'F224A', 'K208A', 'V2A', 'K206A', 'N209A',
             'Q197A', 'R222A', 'G205A', 'E219A', 'E211A', 'D214A']
ax.set_xticks([start_point + 0.1 + i * spacing for i in range(num_points)])
ax.set_xticklabels(x_labels, fontsize=7)


# # 设置y轴标签和字体大小
# ax.set_yticklabels(fontsize=15)

# 设置y轴范围
ax.set_xlim(0, x2[-1])
ax.set_ylim(-3, 3)

# 设置y轴标签
ax.set_ylabel('$\Delta$$\Delta$C(kcal/mol)', fontproperties=myfont, fontsize=19)

# 添加图例
ax.legend(fontsize=7, loc='upper right', frameon=False, prop=myfont, handlelength=2.0)

# 添加水平参考线
# ax.axhline(2, c='b', linewidth=1, ls='--')
# 在每个y轴刻度上添加参考线
for tick in ax.get_yticks():
    ax.axhline(tick, color='black', alpha=0.3, linewidth=0.5)
    
# 隐藏边框
ax.spines['top'].set_color('none')
# ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')

# 隐藏刻度
ax.tick_params(axis='both', which='both', length=0)

# 调整图形布局
plt.subplots_adjust(bottom=0.15, left=0.15)

# 显示图形
plt.show()
```
